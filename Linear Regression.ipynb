## Let's Learn to code

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)

#Reding Data
data = pd.read_csv('headbrain.csv')
print(data.shape)
data.head()
```

```python
#so we can find a linear and chip between the head size
#and the Brain weights.
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values
```

```python
# Mean X and Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

#Total number of values
m = len(X)

#using the formula to calculate b1 and b2
number = 0
denom = 0
for i in range(m):
    number += (X[i]-mean_x)*(Y[i] - mean_y)
    denom += (X[i]-mean_x)**2

b1 = number/denom
b0 = mean_y - (b1 * mean_x)

#Print coefficients
print(b1, b0)
```

```python
#Ploting values and Regression line
max_x = np.max(X)+100
min_x = np.min(X)-100

#Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

#Ploting line
plt.plot(x, y, color='#58b970', label= 'Regression')
#Ploting Scatter Points
plt.scatter(X, Y, c='#ef5423', label= 'Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.show()
```

```python
ss_t = 0
ss_r = 0

for i in range(m):
    y_pred = b0 + b1*X[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)
print(r2)
```

## Linear Regression Multiple Variables

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
df = pd.read_csv('wine-quality.csv', names=["Id","Fixed Acidity","Volatile Acidity","Citric Acid","Residual Sugar","Chlorides","Free sulfur dioxide","Total Sulfur Dioxide","Density","pH","Sulphates","Alcohol","Quality"], header=0)
df.head()
```

```python
df.drop("Id", axis=1, inplace=True)
df.head()
```

```python
X, y = df.drop('Quality',axis=1),df['Quality']
X.head()
```

## Train & Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

### Create to class object

```python
reg = linear_model.LinearRegression()
##Using your training set
reg.fit(X_train,y_train)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
predict = reg.predict(X_test)
```

### Metrics & Evaluated the Model

```python
from sklearn.metrics import r2_score, median_absolute_error, mean_squared_error

r2 = r2_score(y_test, predict)
mae = median_absolute_error(y_test, predict)
mse = mean_squared_error(y_test, predict)
#print("MSE \t MAE \t R2")
print("MSE: "+str(mse)+"\tMAE: "+str(mae)+"\tR2: "+str(r2))
```
